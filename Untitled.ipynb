```python
import kfp
from kfp import dsl
```

```python
def preprocess_op():
    return dsl.ContainerOp(
        name='Preprocess Data',
        image='python:3.8',
        arguments=[
            
        ],
        file_outputs={
            'input':'/tmp/data/res.txt',
            'output':'/tmp/data/output.txt'
        }
    )
```

```python
def train_op(input_param,output_path):
    return dsl.ContainerOp(
        name='write_text',
        image='python:3.8',
        arguments=[
            '--input', input_param,
            '--output', output_path,
        ],
        file_outputs={
            'output': output_path,
        }
    )
```

```python
@dsl.pipeline(
    name='write text pipeline',
    description='读取一个文件写入到另一个文件当中'
)
def read_and_write():
    _preprocess_op = preprocess_op()

    _train_op = train_op(
        dsl.InputArgumentPath(_preprocess_op.outputs['input']),
        dsl.InputArgumentPath(_preprocess_op.outputs['output'])
       
    ).after(_preprocess_op)
```

```python
kfp.compiler.Compiler().compile(read_and_write, 'demo.tar.gz')
```
